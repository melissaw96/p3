```python
gender_dr1 = new_df.Gender_Drv1.value_counts()
gender_dr1
```

```python
gender_dr2 = new_df.Gender_Drv2.value_counts()
gender_dr2
```

```python
gender = new_df.groupby(["Gender_Drv1"])["Gender_Drv2"].count()
labels =["Female","Male","other"]
plt.pie(gender,labels=labels,startangle=180,autopct="%1.1f%%")
plt.title("Gender")
plt.show()
```

```python
fatalities_count = new_df.groupby('Year')['Totalfatalities'].count()
fatalities_count

fatalities_count.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Total Fatalities')
plt.title('Incidents by Year')
plt.show()
```
